```python
import numpy as np
import scipy as sp
from scipy.integrate import nquad
```

```python
G = 6.67e-11
A = 5
a = 1e7
h = 2e3
P = 1e9
rho = 5500
```

```python
func = lambda z,y,x: 1 / (x**2 + y**2 + z**2 + 1e-10)**(0.5)
```

```python
cons = nquad(func, [[0, 1], [-a/h, a/h], [-a/h, a/h]])
```

```python
hunc = lambda z,s,x,y: 1 / (x**2 + y**2 + z**2 + 1e-10)**(0.5)
ranz = lambda s,x,y: (-s, 1-s)
```

```python
res = nquad(hunc, [ranz, [0, 0.5], [-a/h, a/h], [-a/h, a/h]], opts={'epsrel':1e-6})
```

```python
print(res)
```

```python
# h = 2e3
print(2 * res[0] - cons[0], np.pi / 3)
```

```python
# h = 2e4
print(2 * res[0] - cons[0], np.pi / 3)
```

```python
# h = 2e5
print(2 * res[0] - cons[0], np.pi / 3)
```

```python
# h = 2e6
print(2 * res[0] - cons[0], np.pi / 3)
```

```python
end_res = G * (rho**2) * (h**3) * A * (2 * res[0] - cons[0])
```

```python
print("Time in days:", end_res / (P * 3600 * 24))
```
